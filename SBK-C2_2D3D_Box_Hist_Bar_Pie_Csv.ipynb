# Function for reading csv file

```python
# Function to read csv file and return data as 2-d Python list
def read_file_housing(filename):
    file_open = open(filename,"r")
    data_array = []
    for line in iter(file_open):
        line_no_newline = line.rstrip()
        line_split = line_no_newline.split(",")
        data_array.append(line_split) 
    file_open.close()
    return data_array
```

```python
# HW7) Create bar chart with zipcode on x-axis and average_price_by_zipcode on y-axis
```

```python
# Your code here ...
```

```python
def average(lst_of_numbers):
    return float(sum(lst_of_numbers))/len(lst_of_numbers)

def calculate_avg_price_by_zip(arr, priceColumn, zipColumn, zipCode):
    return average([int(x[priceColumn]) for x in arr if int(x[zipColumn]) == zipCode])


ZipCode = extract_column_simple(home_prices, housing_columns.index("ZipCode"))
ZipCode = sorted(set(ZipCode))

AvgPrice = [calculate_avg_price_by_zip(home_prices, housing_columns.index("HomePriceK"), 
                                       housing_columns.index("ZipCode"), x) for x in ZipCode]

print("ZipCode :", ZipCode)
print("AvgPrice", AvgPrice)
```

```python
plt.close()
plt.rcParams.update({'figure.autolayout': True})
fig, ax = plt.subplots(figsize=(10,5))
x = [str(x) for x in ZipCode]
y = list(AvgPrice)
print( x, y )
ax.bar(x, y)
ax.set_title('Bay Area Home Prices - Avg Home Price')
ax.set_xlabel("Zip Code")
ax.set_ylabel("Home Price($K)")
plt.show()
```

```python
# HW9) Create pie chart for home prices
# a) Find the minimum home price, maximum home price, and then split the range into
#    reasonable number of bins (example 6 bins)
# b) Find percentage of homes in each bin and plot the pie chart
```

```python
# Your code here ...
```

```python
def computeBinBoundaries(minPrice, maxPrice, binCount):
    binSize = float(maxPrice - minPrice)/binCount
    price = minPrice
    return [round(price+(x * binSize)) for x in range(binCount+1)]

def computeCountInRange(arr, min, max):
    count = 0
    for i in arr:
        if i >= min and i <= max:
            count += 1
    return count

def computeBinRanges(binBoundaries, arr):
    ranges = {}
    for i in range(len(binBoundaries)-1):
        left = binBoundaries[i]
        right = binBoundaries[i+1]-1 if i < len(binBoundaries)-2 else binBoundaries[i+1]
        rangeLabel = "$" + str(left) + "-" + str(right) + "K"
        rangeValue = computeCountInRange(arr, left, right)
        ranges[rangeLabel] = rangeValue
    return ranges
        

## sortedPrice is sorted in reverse order
print(len(sortedPrice))
minPrice = sortedPrice[len(sortedPrice)-1]
maxPrice = sortedPrice[0]
print(minPrice, maxPrice)

#set bar chart bin counts
binCount = 6

# generate an array of values to be used by binRanges
binBoundaries = computeBinBoundaries(minPrice, maxPrice, binCount)
binRanges = computeBinRanges(binBoundaries, sortedPrice)

print(binBoundaries)
print(binRanges)
print(binRanges.keys())
print(binRanges.values())
print(sum(binRanges.values()))
```

```python
labels = binRanges.keys()
sizes = binRanges.values()
```

```python
plt.close()
plt.rcParams.update({'figure.autolayout': True})
fig, ax = plt.subplots(figsize=(10,5))

ax.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=0)

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax.set_title('Bay Area Home Prices - Price Range Pie Chart')
ax.set_xlabel("Home Price($K)")
plt.show()
```
